28.Generate new data frame for every state (hint: dynamic variable creation) 

```python
grouped_states = df.groupby('State')

state_dataframes = {}
for state, data in grouped_states:
    state_variable_name = f"df_{state.replace(' ', '_')}"  # Replace spaces with underscores for variable names
    state_dataframes[state_variable_name] = data

    # Optionally print the first few rows of each state's DataFrame
    print(f"DataFrame for {state}:\n", data.head())

# Access a specific state's DataFrame by name
print("\nExample: DataFrame for California:")
print(state_dataframes['df_California'].head())
```
